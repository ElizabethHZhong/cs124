```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
d = pd.read_csv('timing-data.csv')

print(d.head())
```

```python
fig, ax = plt.subplots()
sns.stripplot(data=d, x="Function", y="Residue", hue="Function", ax=ax)
fig.set_figwidth(12)
fig.set_figheight(6)
ax.set_yscale('log')
ax.set_title("Residue Scatterplot")
plt.savefig('egg.png')
```
